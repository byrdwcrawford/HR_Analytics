# Predicting Employee Attrition Rate  

### Overview  

This project sets out to determine attrition rate for employees in a company.  As we have seen an adoption of hiring and firing quickly within organizations across the US, it is vitally important for companies to be able to predict attrition rates of employees.  This can be helpful in resource planning and also saving companies money as it is expensive to hire and fire.  

Personally, I come from the recruitment industry.  I have worked in Healthcare and IT recruitment and have seen first hand how costly attrition can be.  Losing employees is expensive.  Best case scenario is an organizatoin can lean on a trusted staffing vendor to supply consultants.  However, often times there is a trickle down effect of attrition.    

This project will help organizations anticipate attrition, which in turn, allows the company to better strategize in the following ways:  

* Resource Planning
* Cost Saving
* Maintaining Productivity
* Counter measures to prevent employees leaving



### Data Understanding

HR Analytics data was collected [here](https://www.kaggle.com/competitions/house-prices-advanced-regression-techniques/data).  This is a 1479 x 35 publicly available CSV.  Important features are going to be our target value **'Attrition'**, and various features such as:  

* Job Satisfaction
* Age
* Sex
* Job Title
* Salary

### Evaluation

The target variable is **Attrition** and more specifically, we will be targeting **recall** for this class as we want to correctly identify all positive cases in our dataset.  If our model misses an employee who is likely to leave, we lose an employee without taking preventative actions.  

The MVP will be a Decision Tree Classifier with class imbalanced addressed via class weights, SMOTE, and Undersampling with stratification.  

To improve upon this MVP, Ensemble methods will be used, unimportant features will be dropped, and all models will be finetuned.  The trade-off between recall of the minority (target) class and overall accuracy, F1 score, Confusion Matrix, and ROC-AUC will ultimately determine what our final model will look like.  Featuer Importances will provide context
